```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bottom_left = cv2.imread('bottom_left.jpg')
bottom_left = cv2.cvtColor(bottom_left,cv2.COLOR_BGR2RGB)
crop_bl = cv2.resize(bottom_left,(200,200))
border_bl = cv2.copyMakeBorder(crop_bl,5,10,10,5,cv2.BORDER_CONSTANT)
plt.imshow(border_bl)
border_bl.shape
```

```python
bottom_right = cv2.imread('bottom_right.jpg')
bottom_right = cv2.cvtColor(bottom_right,cv2.COLOR_BGR2RGB)
crop_br = cv2.resize(bottom_right,(200,200))
border_br = cv2.copyMakeBorder(crop_br,5,10,5,10,cv2.BORDER_CONSTANT)
plt.imshow(border_br)
print(border_br.shape)
```

```python
top_left = cv2.imread("top_left.jpg")
top_left = cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB)
crop_tl = cv2.resize(top_left,(200,200))
border_tl =cv2.copyMakeBorder(crop_tl,10,5,10,5,cv2.BORDER_CONSTANT)
plt.imshow(border_tl)
border_tl.shape
```

```python
top_right = cv2.imread('top_right.jpg')
top_right = cv2.cvtColor(top_right,cv2.COLOR_BGR2RGB)
crop_tr = cv2.resize(top_right,(200,200))
border_tr = cv2.copyMakeBorder(crop_tr,10,5,5,10,cv2.BORDER_CONSTANT)
plt.imshow(border_tr)
border_tr.shape
```

```python
center = cv2.imread('center.jpeg')
center = cv2.cvtColor(center,cv2.COLOR_BGR2RGB)
crop_center = cv2.resize(center, (100,100))
border_center  = cv2.copyMakeBorder(crop_center,10,10,10,10,cv2.BORDER_CONSTANT)
plt.imshow(border_center)
x,y,z = border_center.shape
print(x,y,z)
```
